```python
import pandas as pd
import numpy as np
```

```python
spreadsheet = pd.read_excel('COVID-19-geographic-disbtribution-worldwide.xlsx')
cols_spreadsheet = spreadsheet[['day', 'month', 'year', 'cases', 'deaths', 'countriesAndTerritories', 'countryterritoryCode', 'popData2019']]
spreadsheet
```

```python
aggregated_deaths_and_cases_since_sept = cols_spreadsheet.where(cols_spreadsheet['month'] >= 9).groupby(['countriesAndTerritories', 'countryterritoryCode', 'popData2019'], as_index=False)[['cases', 'deaths']].sum().rename(columns={'cases': 'cases_since_sept', 'deaths': 'deaths_since_sept'})
aggregated_deaths_and_cases_since_sept
```

```python
aggregated_deaths_and_cases_since_oct = cols_spreadsheet.where(cols_spreadsheet['month'] >= 10).groupby(['countriesAndTerritories', 'countryterritoryCode', 'popData2019'], as_index=False)[['cases', 'deaths']].sum().rename(columns={'cases': 'cases_since_oct', 'deaths': 'deaths_since_oct'})
aggregated_deaths_and_cases_since_oct
```

```python
dataset = aggregated_deaths_and_cases_since_sept.join(aggregated_deaths_and_cases_since_oct[['cases_since_oct', 'deaths_since_oct']])
dataset['percent_of_oct_cases_in_oct_and_sept'] = dataset['cases_since_oct'] / dataset['cases_since_sept']
dataset['percent_of_oct_deaths_in_oct_and_sept'] = dataset['deaths_since_oct'] / dataset['deaths_since_sept']
dataset
```

```python
per100k_dataset = dataset
per100k_dataset['cases_since_sept_per100k'] = dataset['cases_since_sept'] / dataset['popData2019'] * 100000.0
per100k_dataset['deaths_since_sept_per100k'] = dataset['deaths_since_sept'] / dataset['popData2019'] * 100000.0
per100k_dataset['cases_since_oct_per100k'] = dataset['cases_since_oct'] / dataset['popData2019'] * 100000.0
per100k_dataset['deaths_since_oct_per100k'] = dataset['deaths_since_oct'] / dataset['popData2019'] * 100000.0
per100k_dataset['countryterritoryCode'] = per100k_dataset['countryterritoryCode'].astype(str)
```

```python
def find_last_valid_value_in_row(row):
    if row.last_valid_index() is None or type(row[row.last_valid_index()]) == str:
        return np.nan
    else:
        return row[row.last_valid_index()]
```

```python
temps = pd.read_excel('cckp_historical_data_0.xls', sheet_name='Country_temperatureCRU')[['ISO_3DIGIT', 'Sept_temp', 'Oct_temp']]
temps = temps.rename(columns={'ISO_3DIGIT': 'countryterritoryCode'})
per100k_dataset = per100k_dataset.join(temps.set_index('countryterritoryCode'), on='countryterritoryCode')
```

```python
gdp_per_capita = pd.read_excel('API_NY.GDP.PCAP.CD_DS2_en_excel_v2_1678737.xls', sheet_name='Data', skiprows=3)
last_recorded_gdp = gdp_per_capita[[str(year) for year in range(1960, 2020)]]
gdp_per_capita['gdp_per_capita'] = last_recorded_gdp.apply(find_last_valid_value_in_row, axis=1)
gdp_per_capita = gdp_per_capita[['Country Code', 'gdp_per_capita']]
gdp_per_capita = gdp_per_capita.rename(columns={'Country Code': 'countryterritoryCode'})
per100k_dataset = per100k_dataset.join(gdp_per_capita.set_index('countryterritoryCode'), on='countryterritoryCode', how='left')
per100k_dataset
```
